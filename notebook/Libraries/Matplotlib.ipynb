```python
plt.figure(figsize=(10,6), dpi=80)

ax = plt.subplot(111)
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_color('none')
ax.spines['left'].set_color('none')
ax.yaxis.set_ticks_position('left')

plt.plot(x, C,  color='blue', linewidth=3.0, linestyle="-", label="cosine")
plt.plot(x, S, color='red', linewidth=3.0, linestyle="-", label="sin")

## set x limits
plt.xlim(-4.0, 4.0)

# set x ticks
# latex formating of scientific symbols
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi],
          [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])

# set y limits
plt.ylim(-1.0, 1.1)

# y ticks
plt.yticks(np.linspace(-1, 1, 6, endpoint=True))

t = 2 * np.pi / 3
plt.plot([t,t], [0, np.cos(t)],  color='blue', linewidth=3.0, linestyle="-")
plt.scatter([t,  ], [np.cos(t), ], 50, color='blue')
plt.annotate(r'$\sin(\frac{2\pi}{3})=\frac{\sqrt{3}}{2}$',
                    xy=(t, np.sin(t)), xycoords='data',
                    xytext=(+15, +10), textcoords='offset points', fontsize=14,
                    arrowprops=dict(arrowstyle="->", connectionstyle="arc3, rad=.2"))

plt.plot([t,t], [0, np.sin(t)], color='red', linewidth=3.0, linestyle="-")
plt.scatter([t,  ], [np.sin(t), ], 50, color='red')

plt.annotate(r'$\cos(\frac{2\pi}{3})=-\frac{1}{2}$',
                    xy=(t, np.cos(t)), xycoords='data',
                    xytext=(-90, -50), textcoords='offset points', fontsize=14,
                    arrowprops=dict(arrowstyle="->", connectionstyle="arc3, rad=.2"))
# set legend
plt.legend(loc='upper left', frameon=False)

plt.show()
```

#### Bar plot

```python
import numpy as np
import matplotlib.pyplot as plt


#create sample dat set
data = {'C':20,'C++':15,'Java':30,'Python':45}

courses = list (data.keys())
values = list(data.values())

fig = plt.figure(figsize=(10,6))

#create the plot
plt.bar(courses, values,color='pink',width=0.5)
plt.xlabel("our courses")
plt.ylabel("no of session")
plt.title("our awesome matplotlib")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt


#create sample dat set
data = {'C':20,'C++':15,'Java':30,'Python':45}

courses = list (data.keys())
values = list(data.values())

fig = plt.figure(figsize=(10,6))

#create the plot(horizontal bar)
plt.barh(courses, values,color='olive')
plt.xlabel("our courses")
plt.ylabel("no of session")
plt.title("our awesome matplotlib")
plt.show()
```

```python
import pandas as pd

data = pd.read_csv("../cars.csv")
data
```

```python
name = data['car'].head(10)
hp =data['hp'].head(10)

fig = plt.figure(figsize = (10,8))
plt.bar(name,hp,color='violet')
plt.show()
```

```python
name = data['car'].head(10)
hp =data['hp'].head(10)

fig,ax= plt.subplots(figsize = (16,9))

#horizontal bar plot
ax.barh(name,hp)

#remove plot spines
for side in ['top','bottom','left','right']:
    ax.spines[side].set_visible(True)

#remove x,y ticks
ax.xaxis.set_ticks_position('none')
ax.yaxis.set_ticks_position('none')

#add padding
ax.xaxis.set_tick_params(pad=5)
ax.yaxis.set_tick_params(pad=5)

#add grid lines
ax.grid(b=True,color='black',linestyle='-.',linewidth=0.5,alpha=0.2)

ax.invert_yaxis()

#add annotations
for i in ax.patches:
    plt.text(i.get_width()+0.9,i.get_y()+0.5,str(round(i.get_width(),2)),
             fontsize=12,fontweight='bold',color='grey')

#add watermark
fig.text(0.89,0.15,'Varun S',fontsize=16,color='black',ha='right',va='bottom',alpha=1)

ax.set_title('Students with their usage of phones')

#plt.bar(name,lapusg)
plt.show()
```

#### Scatterplot

```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use("seaborn-whitegrid")
```

```python
x = np.linspace(0,10,30)
y=np.sin(x)

plt.plot(x,y,'o',color='black')
```

```python
rng = np.random.RandomState(0)
for marker in ['o','.',',','x','+','v','^','>','<','s','d']:
    plt.plot(rng.rand(5),rng.rand(5),marker,label = "marker='{0}'".format(marker))
    
plt.legend(numpoints = 1)
plt.xlim(0,1.8)
```

```python
x = np.linspace(0,10,30)
y=np.sin(x)

plt.plot(x,y,'o',color='black')
plt.plot(x,y,'-ok',color='red')
```

```python
plt.plot(x,y,'-p',color='red',
        markersize = 15,linewidth=2,
        markerfacecolor='yellow',
        markeredgecolor='black',
        markeredgewidth=2)

plt.ylim(-1.2,1.2)
```

#### Bubble plot

```python
rng = np.random.RandomState(7)
x = rng.randn(100)
y = rng.randn(100)
colors = rng.randn(100)
sizes = 1000*rng.randn(100)

plt.scatter(x,y,c=colors,s=sizes,alpha=0.4,cmap='viridis')
plt.colorbar()
```
