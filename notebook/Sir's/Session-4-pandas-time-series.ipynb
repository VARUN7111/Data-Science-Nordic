## Resampling, Shifting, windowing

```python
D  = day
W = week

```

```python
df = pd.DataFrame({
    "date": pd.date_range(start='2021-03-01', periods=100, freq='D'),
    # create a array of integers valued between 18-40,
    # create an array on floats with decimal places recuded to one digit
    "temperature": np.random.randint(18, 40, size=100) + np.random.random(100).round(1)
})
df.head()
```

```python
np.random.randn(100).round(1)
```

```python
np.random.randint(18, 40, size=100)
```

```python
df.tail()
```

 ## Shift 

```python
df['lag'] = df['temperature'].shift(1)
df.head()
```

```python
df['lag']  = df['temperature'].shift(1, fill_value=df['temperature'].mean())
df.head()
```

### Resample

```python
df_weekly = df.resample('W', on='date').mean()
df_weekly.head()
```

```python
df_m = df.resample('M', on='date').mean()
df_m.head()
```
