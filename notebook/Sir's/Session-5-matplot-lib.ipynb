## Bar plot

```python
import numpy as np
import matplotlib.pyplot as plt


# create sample dataset
data =  {'C': 20, 'C++': 15, 'Java': 30, 'Python': 45}

courses = list(data.keys())
values = list(data.values())

fig = plt.figure(figsize=(10, 6))

# create the plot
plt.bar(courses, values , color='lightblue', width=0.6)
plt.xlabel('Our cources')
plt.ylabel('Number of sessions')
plt.title('Our awesome matplotlib')
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('../cars.csv')
data.head()
name = data['car'].head(10)
hp = data['hp'].head(10)

fig, ax = plt.subplots(figsize=(16, 9))

# horizontal bar plot
ax.barh(name, hp)

# remove plot spines
for side in ['top', 'bottom', 'left', 'right']:
    ax.spines[side].set_visible(False)
    
# Remove    x, y ticks
ax.xaxis.set_ticks_position('none')
ax.yaxis.set_ticks_position('none')

#  add padding
ax.xaxis.set_tick_params(pad=5)
ax.yaxis.set_tick_params(pad=8)

# Add grid lines
ax.grid(b=True, color='gray',linestyle='-.', linewidth=0.5, alpha=0.2)

ax.invert_yaxis()

# Add some annotations
for i in ax.patches:
    plt.text(i.get_width() + 0.2, i.get_y() + 0.5, str(round((i.get_width()), 2)),
             fontsize= 12, fontweight='bold', color='gray')

ax.set_title('sports cars with horse power', loc='left')

# add watermark
fig.text(0.9, 0.15, 'Nordic Academy Bar Plot', fontsize=16, color='lightgray', ha='right', va='bottom', alpha=0.5)
plt.show()
```

## Scattorplots

```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
```

```python
x = np.linspace(0, 10, 30)
y = np.sin(x)

plt.plot(x, y, 'o', color='black')
```

```python
rng = np.random.RandomState(0)
for marker in ['o', '.', ',', 'x', '+', 'v', '^', '<', '>', 's', 'd']:
    plt.plot(rng.rand(5), rng.rand(5), marker, label="marker='{0}'".format(marker))

plt.legend(numpoints=1)
plt.xlim(0, 1.8)
```

```python
plt.plot(x, y, '-ok')
```

```python
plt.plot(x, y, '-p', color='gray',
        markersize=15, linewidth=4,
        markerfacecolor='white',
        markeredgecolor='gray',
        markeredgewidth=2)
plt.ylim(-1.2, 1.2)
```

## Bubble plots

```python
rng = np.random.RandomState(7)
x = rng.randn(100)
y = rng.randn(100)
colors = rng.randn(100)
sizes = 1000 * rng.randn(100)

plt.scatter(x, y, c=colors, s=sizes, alpha=0.3, cmap='viridis')
plt.colorbar()
```
